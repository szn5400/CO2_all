```python
import matplotlib.pyplot as plt
def display_permeability_data(timestamp, input):
  plot_x_target = [[0 for i in range(48)] for j in range(54)]
  plot_y_target = [[0 for i in range(48)] for j in range(54)]
  plot_z_target = [[0 for i in range(48)] for j in range(54)]

  for i in range(54):
      for j in range(48):
        for k in range(3):
          #plot_x_target[i][j] = (input[1][timestamp-1][i][j][0])
          #plot_y_target[i][j] = (input[1][timestamp-1][i][j][50])
          #plot_z_target[i][j] = (input[1][timestamp-1][i][j][90])
          plot_x_target[i][j] = np.log10(input[0][timestamp][i][j][0])
          plot_y_target[i][j] = np.log10(input[0][timestamp][i][j][1])
          plot_z_target[i][j] = np.log10(input[0][timestamp][i][j][81])

  plt.imshow(np.squeeze(plot_x_target), cmap="jet")
  plt.title("z=1")
  plt.colorbar()
  #plt.clim(-0.5,2.0)
  plt.show()
  

  plt.imshow(np.squeeze(plot_y_target), cmap="jet")
  plt.title("z=2")
  plt.colorbar()
  #plt.clim(-0.5,2.0)
  plt.show()

  plt.imshow(np.squeeze(plot_z_target), cmap="jet")
  plt.title("z=3")
  plt.colorbar()
  #plt.clim(-0.5,2.0)
  plt.show()
```
